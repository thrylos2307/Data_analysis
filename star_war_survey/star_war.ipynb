## Star Wars: The Force Awakens 

" Does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?"

To answer this question we have data from survey of Star Wars fans through SurveyMonkey.
The data has the encoding: "ISO-8859-1"

Coloumns data contains :
- RespondentID - An anonymized ID for the respondent (person taking the survey)
- Gender - The respondent's gender
- Age - The respondent's age
- Household Income - The respondent's income
- Education - The respondent's education level
- Location (Census Region) - The respondent's location
- Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
- Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response

Some of the row are blank ,so let's start with data cleaning.

RespondentID is supposed to be a unique ID for each respondent, but it's blank in some rows.


```python
import pandas as pd 
import numpy as np
```

```python
star_wars=pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

##### Columns 

```python
star_wars.columns
```

Deleting row that contains Nan in ResponseId

```python
star_war=star_wars[star_wars['RespondentID'].notnull()]
```

```python
star_war
```

If we look at columns:
- Have you seen any of the 6 films in the Star Wars franchise?
- Do you consider yourself to be a fan of the Star Wars film franchise?

Both represent Yes/No questions. They can also be NaN where a respondent chooses not to answer a question

So let's look at the how time Nan value occurs. For the ease lets convert the data to True and False value through mapping.

```python
yes_no = {
    "Yes": True,
    "No": False
}
for col in [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]:
    star_war[col] = star_war[col].map(yes_no)

star_war.head()
```

Next are some columns with checbox questions .
The columns for this question are:

* Which of the following Star Wars films have you seen? Please select all that apply. - Whether or not the respondent saw Star Wars: Episode I The Phantom Menace.
* Unnamed: 4 - Whether or not the respondent saw Star Wars: Episode II Attack of the Clones.
* Unnamed: 5 - Whether or not the respondent saw Star Wars: Episode III Revenge of the Sith.
* Unnamed: 6 - Whether or not the respondent saw Star Wars: Episode IV A New Hope.
* Unnamed: 7 - Whether or not the respondent saw Star Wars: Episode V The Empire Strikes Back.
* Unnamed: 8 - Whether or not the respondent saw Star Wars: Episode VI Return of the Jedi.

These cell contains the response that he saw that movie or have Nan value which means they don't see movie.
 For working with data we need to convert these columns to Boolean as we done before 

```python
movie_map={
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}
```

```python
for i in star_wars.columns[3:9]:
    star_war[i]=star_war[i].map(movie_map)
```

```python
col_name={
    "Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
    "Unnamed: 4":"seen_2",
    "Unnamed: 5":"seen_3",
    "Unnamed: 6":"seen_4",
    "Unnamed: 7":"seen_5",
    "Unnamed: 8":"seen_6",
}
```

```python
star_war=star_war.rename(columns=col_name)
```

```python
star_war.head()
```

Next six columns ask the rating for the movies.1 means the film was the most favorite, and 6 means it was the least favorite. Each of the following columns can contain the value 1, 2, 3, 4, 5, 6, or NaN

```python
star_war[star_war.columns[9:15]] = star_war[star_war.columns[9:15]].astype(float)
```

```python
star_war.columns[9:15]
```

```python
ranking_rename={
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
       'Unnamed: 10':'ranking_2',
    'Unnamed: 11':'ranking_3',
    'Unnamed: 12':'ranking_4',
    'Unnamed: 13':'ranking_5',
       'Unnamed: 14':'ranking_6'
}
star_war=star_war.rename(columns=ranking_rename)
```
